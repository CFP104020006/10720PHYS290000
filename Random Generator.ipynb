## Random Generator: Uniform

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.random.random(100)
print(x)
```

## histogram

```python
plt.hist(x, 10)

plt.xlabel("x value")
plt.ylabel("Numbers of elements")
plt.show
```

## Random Generator: Gaussian

```python
mu, sigma = 0, 0.1#mean and standard deviation
s = np. random.normal (mu, sigma, 100000)
plt.hist(s, 30)
#plt.xlim(-2, 2)
plt.show()
```

```python
ss = plt.hist(s, 10)
print(ss)
```

```python
bins = np.linspace(-0.5, 0.5, 21, endpoint=True)
print(bins)
plt.hist(s, bins)
plt.xlim(-2, 2)
plt.show()
```

## Add Noise To Black Body Radiation

```python
import numpy as np
import matplotlib.pyplot as plt
import math

h = 6.62*10**-34
c = 3*10**8
k = 1.38*10**-23

num = 100

#T = float(input("input Temperature:"))
T = 5000
v = np.linspace(10**0, 10**15, num, endpoint=True)
B = 2*h*v**3/c**2/(math.e**(h*v/k/T-1))


# plt.xlim(10**15)
#plt.plot(v, B,"r.")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Intensity Watt/Hz/m^2")

dB = np.random.normal(0., T*10**-12, num)
B += dB
plt.plot(v,B)
plt.show()
```
